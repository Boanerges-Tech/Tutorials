## Covid Package

```python
from covid import Covid
```

### Get all The Data

```python
covid = Covid()
print(covid.get_data())
```

### Get all The Countries who are affected to the Covid-19

```python
countries = covid.list_countries()
print(countries)
```

### Get the Status by Country Name or id

```python
germany_cases = covid.get_status_by_country_name("Germany")
print(germany_cases)
```

### How to get total Active Cases/ Recovered Cases/ number of Deaths

```python
active = covid.get_total_active_cases()
print(active)
```

```python
recover = covid.get_total_recovered()
print(recover)
```

```python
death = covid.get_total_deaths()
print(death)
```

### Getting Data from The Worldometer

```python
worldometer = Covid(source="worldometers")
print(worldometer.get_data())
```

## COVID19Py Package

```python
import COVID19Py
```

```python
covid19 = COVID19Py.COVID19(data_source="jhu")
print(covid19.getLatest())
```

```python
print(covid19.getLocations())
```

### Rank the Results

```python
confirmed_cases = covid19.getLocations(rank_by='confirmed')
print(confirmed_cases)
```

### Get by Country Code

```python
location = covid19.getLocationByCountryCode("IN")
print(location)
```
